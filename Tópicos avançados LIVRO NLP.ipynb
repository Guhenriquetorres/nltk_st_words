```python
#A ideia é fazer o loop pra todas as páginas
inteiro_teor = ''
for pages in arquivo.pages:
    texto_pagina = pages.extract_text()
    inteiro_teor = inteiro_teor + texto_pagina
```

```python
print(inteiro_teor)
```

```python
inteiro_teor_lower = inteiro_teor.lower()
print(inteiro_teor_lower)
```

```python
# AGORA QUE JÁ FOI REALIZADO A AQUISISÇÃO DO DOCUMENTO FAZER AS ESTATÍSTICAS DAS PALAVRAS 
# COMANDO VIGENTE É PARA ELIMINAR OS NÚMEROS DO PDF 
import re 

inteiro_teor_lower = re.sub(r'\d','',inteiro_teor_lower)
print(inteiro_teor_lower)
```

```python
# [^\w\s] Significa qualquer caractere que não seja: \w: letras, números ou underscore (_).\s: espaços ou quebras de linha.
#Esta linha remove todos os caracteres especiais e pontuações, como .,;:!?()[] etc.



inteiro_teor_lower = re.sub(r'[^\w\s]','', inteiro_teor_lower )



#'\s+' Corresponde a um ou mais caracteres de espaço (espaços, tabs, quebras de linha).
#Substitui essas sequências por um único espaço simples ' '.
#Isso padroniza o espaçamento e deixa o texto mais limpo e uniforme.

inteiro_teor_lower = re.sub(r'\s+', ' ', inteiro_teor_lower)
print(inteiro_teor_lower)
```

```python
print(inteiro_teor_lower)
```

```python
# JA TRADADO O PDF APENAS COM PALAVRAS, É IMPORTANTE RELEMBRAR QUE SERÁ AVALIADO APENAS AS ESTATÍSTICAS E NÃO O SENTIDO
# UTILIZAREMOS EXPRESSÕES REGURARES COM REGEX

# PROCURA POR TODOS OS VALORES SEM ESAÇOS, EM SEGUIDA MONTA O ARRAY COM AS PALAVRAS

regex_token = r'\W+'
inteiro_teor_lower_token = re.split(regex_token, inteiro_teor_lower)
```
